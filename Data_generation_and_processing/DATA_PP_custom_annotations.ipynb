COPYING FILES TO MAKE ONE COMBINED FOLDER

```python
folder1 = r"C:\AAKASH\MS_NOTES\6th sem\RAML\extracted_frames_aug_jpg"
folder2 = r"C:\AAKASH\MS_NOTES\6th sem\RAML\extracted_frames_jpg"
combined_folder = r"C:\AAKASH\MS_NOTES\6th sem\RAML\combined_images"  # Destination folder

if os.path.exists(combined_folder):
    shutil.rmtree(combined_folder)
os.makedirs(combined_folder, exist_ok=True)

def copy_all(src_dir, dst_dir):
    for file_name in os.listdir(src_dir):
        src_path = os.path.join(src_dir, file_name)
        dst_path = os.path.join(dst_dir, file_name)
        if os.path.isfile(src_path):
            shutil.copy2(src_path, dst_path)

copy_all(folder1, combined_folder)
copy_all(folder2, combined_folder)
```

SETTING UP PATHS 

```python
image_dir = r"C:\AAKASH\MS_NOTES\6th sem\RAML\combined_images"
output_dir = r"C:\AAKASH\MS_NOTES\6th sem\RAML\extracted_with_annotations"
os.makedirs(output_dir, exist_ok=True)
out_img_dir = os.path.join(output_dir, "images")
out_lbl_dir = os.path.join(output_dir, "labels")

os.makedirs(out_img_dir, exist_ok=True)
os.makedirs(out_lbl_dir, exist_ok=True)
```

INFERENCE AND ANNOTATION

```python
#runs inference using a YOLOv8 model
model = YOLO('yolov8n.pt')
results = model.predict(
    source=image_dir,
    stream=True,                                                        #streaming: low RAM usage
    conf=0.25                                                           #confidence threshold  i.e ignore detections below 25%
)

for r in results:                                                       #loop through predictions
    img_path = r.path
    img_name = os.path.basename(img_path)
    img_stem = os.path.splitext(img_name)[0]

    boxes = r.boxes
    new_lines = []
    for box in boxes:
        cls_id = int(box.cls[0])                                        #COCO class ID
        if cls_id == 15:                                                #COCO cat
            cls_id = 0                                                  #custom cat class
            xywh = box.xywhn[0].tolist()                                #normalized xywh
            xywh = [str(x) for x in xywh]                               #converts to a string
            new_lines.append(f"{cls_id} " + " ".join(xywh))

    if new_lines:
        # If at least 1 cat: copy image + save label
        dst_img_path = os.path.join(out_img_dir, img_name)
        shutil.copy2(img_path, dst_img_path)

        with open(os.path.join(out_lbl_dir, img_stem + ".txt"), 'w') as f:
            for line in new_lines:
                f.write(line + '\n')

    #If no cat: do NOTHING i.e skip image, skip label.
```
